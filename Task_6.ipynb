### Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as dataset
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn import tree
from seaborn import heatmap
```

### Importing Data and Preprocessing Data

```python
iris = dataset.load_iris()
```

```python
xdata = pd.DataFrame(iris.data,columns=iris.feature_names)
print(xdata.head(5))
ydata = iris.target
print(ydata[:5])
```

```python
xdata.describe()
```

```python
xdata.isna().sum()
```

### Visualising the correlation among different values and categories in data

```python
heatmap(xdata)
```

### Creating a decision tree model

```python
xtrain, xtest, ytrain, ytest = train_test_split(xdata,ydata,test_size=0.2)
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(xtrain,ytrain)
print('Decision Tree Created!!!')
```

### Visualising Decision Tree

```python
fig,axis = plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300)
tree.plot_tree(decision_tree,feature_names=xdata.columns,filled=True);
fig.savefig('Dataset/Decision Tree.png')
```

### Testing the model created

```python
y_pred = decision_tree.predict(xtest)
y_pred
```

```python
confusion_matrix(ytest,y_pred)
```

```python
accuracy_score(ytest,y_pred)
```

Accuracy obtained is 90% as the data used for training and testing is not very large and is highly correlated!!!
